```python
lists = df2["listed_in"].str.split(",").copy()
```

```python
new_lists = []
for item in lists:
    new_lists.append(item)

```

```python
new_lists
```

```python

flat_list = [x.strip() for sub in lists for x in sub]
```

```python
flat_list
```

```python
unique_list = {}
for i in range(len(flat_list)):
    count = 0
    for j in range(1 , len(flat_list)):
        if flat_list[i] == flat_list[j]:
            count += 1
    if flat_list[i] not in unique_list:
        unique_list[flat_list[i]] = count



    
```

```python
unique_list
```

```python
sorted_dict = dict(sorted(unique_list.items() , key= lambda x : x[1] , reverse= True ))
sorted_dict
```

```python
top_10 = {}
for i, (k, v) in enumerate(sorted_dict.items()):
    if i == 10:
        break
    top_10[k] = v
```

```python
top_10
```

```python
top_10 = {
    'International Movies': 2752,
    'Dramas': 2427,
    'Comedies': 1674,
    'International TV Shows': 1351,
    'Documentaries': 868,
    'Action & Adventure': 859,
    'TV Dramas': 763,
    'Independent Movies': 756,
    'Children & Family Movies': 641,
    'Romantic Movies': 616
}
top_10df = pd.DataFrame(list(top_10.items()),columns=['Category', 'Count'])
```

```python
import matplotlib.pyplot as plt

categories = ['International Movies', 'Dramas', 'Comedies', 'International TV Shows',
              'Documentaries', 'Action & Adventure', 'TV Dramas', 'Independent Movies',
              'Children & Family Movies', 'Romantic Movies']
counts = [2752, 2427, 1674, 1351, 868, 859, 763, 756, 641, 616]


plt.figure(figsize=(12, 6))
bars = plt.bar(categories, counts, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.title('Top 10 Most Common Netflix Genres')
plt.tight_layout()


for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 20, str(height),
             ha='center', va='bottom', fontsize=9)

plt.show()
```
